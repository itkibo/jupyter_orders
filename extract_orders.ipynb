```python
import json
with open('getorderbook.json', 'r') as file:
    data = json.load(file)
    print(data)
```

```python
# buyers orders
data['bids']
```

```python
# sellers orders
data['asks']
```

```python
# function converts price from quotation to float
def Quotation_to_float(units, nano):
    if not units:
        units = 0
    if not nano:
        nano = 0
    return int(units) + int(nano)/1000000000
```

```python
orders_bids = []

# get bids + format datatype
for row in data['bids']:
    orders_bids.append({
        'bids.quantity': int(row['quantity']),
        'bids.price': Quotation_to_float(row['price']['units'], row['price']['nano'])
    })

orders_bids
```

```python
orders_asks = []

# get asks + format datatype
for row in data['asks']:
    orders_asks.append({
        'asks.quantity': int(row['quantity']),
        'asks.price': Quotation_to_float(row['price']['units'], row['price']['nano'])
    })

orders_asks
```

```python
# make dataframes
import pandas as pd
bids = pd.DataFrame(orders_bids)
asks = pd.DataFrame(orders_asks)
```

```python
bids.info()
```

```python
# join together horizontally
orders =pd.concat(
    objs = [bids, asks],
    axis = 'columns',
    join = 'inner'
)
```

```python
orders
```

```python
# check data types
orders.info()
```

```python
# list of buckets to buy
bids_buckets = [40, 30, 20, 10]
# get n largests from orders bids by quantity
bids_largest = orders['bids.quantity'].nlargest(n = len(bids_buckets))
bids_largest
```

```python
bids_largest.info()
```

```python
bids_largest = pd.DataFrame(bids_largest)
bids_largest
```

```python
bids_largest.info()
```

```python
# make dataframe with indexes from bids largest based on bids buckets data
bids_largest = pd.DataFrame(data=bids_buckets, index=bids_largest.index, columns=['bids.position'])
bids_largest
```

```python
orders = orders.merge(bids_largest, how="left", left_index=True, right_index=True)
orders
```

```python
# list of buckets to buy
asks_buckets = [40, 30, 20, 10]
asks_largest = orders['asks.quantity'].nlargest(n = len(asks_buckets))
asks_largest = pd.DataFrame(data=asks_buckets, index=asks_largest.index, columns=['asks.position'])
asks_largest
```

```python
orders = orders.merge(asks_largest, how='left', left_index=True, right_index=True)
orders
```

```python
# replace nan with zero
orders['bids.position'] = orders['bids.position'].fillna(0)
orders['asks.position'] = orders['asks.position'].fillna(0)
orders
```

```python
orders['bids.position'] = orders['bids.position'].astype(int)
orders['bids.position'] = orders['bids.position'].astype(int)
orders
```

```python
# I interested in four most right pointes on bids quantity axis
# it will be my possible positions in orders boo

from matplotlib import pyplot as plt
orders.plot(kind='scatter', x='bids.quantity', y='bids.price', s=32, alpha=.8)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.grid(True)
```

```python
# Now I interested in four most right pointes on asks quantity axis
# it will be my possible positions in orders boo

from matplotlib import pyplot as plt
orders.plot(kind='scatter', x='asks.quantity', y='asks.price', s=32, alpha=.8)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.grid(True)
```
